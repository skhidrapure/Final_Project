```python
# Import the requests library.
import requests
import pandas as pd
from datetime import datetime
import time
import os

from config import polygon_api_key
```

```python
# Import health stock list
SPX_Health = pd.read_csv("SPX_Health_Sector_info.csv")
SPX_Health.head()
```

```python
SPX_Tickers=SPX_Health['Ticker']
SPX_Tickers
```

```python
Health_stocks = pd.DataFrame(index=None)
Health_stocks
```

```python
i=1
for ticker in SPX_Tickers:
    # URL for Polygon API Call - from 2020-03-12 to 2021-12-31
    url = "https://api.polygon.io/v2/aggs/ticker/" + ticker + "/range/1/day/2020-03-12/2021-12-31?adjusted=true&sort=asc&limit=730&apiKey=" + polygon_api_key
    resp = requests.get(url)
    
    if resp.status_code != 200:
        print(f"Error {url}")
        continue
    
    data=resp.json()
    
    df=pd.DataFrame(data['results'])
    df['Ticker']=data['ticker']
    
    Health_stocks=Health_stocks.append(df)
    # Free version only allows 5 API calls per minute
    time.sleep(20)
    
    # Get updated on the process
    print(f'Getting stock #{i}')
    i=i+1
    
    next

print('Complete')
```

```python
timestamp=Health_stocks['t']/1000
        
date=[]

for i in timestamp:
        time_date=datetime.fromtimestamp(int(i)).strftime("%Y-%m-%d")
        date.append(time_date)
        next
        
Health_stocks['Date']=date
    
```

```python
Columns={'v':'Trading Volume', 'vw':'Volume Weighted Average Price', 'o':'Open', 'c':'Close', 'h':'High', 'l':'Low', 'n':'Number of Transactions'}
Health_stocks=Health_stocks.rename(columns=Columns)
```

```python
Health_stocks=Health_stocks[['Ticker', 'Date', 'Open', 'High', 'Low', 'Close', 'Trading Volume', 'Volume Weighted Average Price', 'Number of Transactions']]
Health_stocks.head()
```
