```python
import pandas as pd
meta = pd.read_csv("datasets/Covid_Companies_Stocks.csv", index_col=False)
meta
```

```python
df = meta[['Ticker', 'Date', 'Close', 'Open']]
df.head()
```

```python
df_pivot = df.pivot(index=['Date'], columns = ['Ticker'], values = ['Close','Open'])
df_pivot.head()
```

## Overview of closing for vaccination companies

```python
close = df_pivot.Close
```

```python
open = df_pivot.Open
```

```python
df_pivot_movement = close - open
df_pivot_movement
```

```python
# Loading S5HLTH Index file.

s5health = pd.read_csv("datasets/sp500indexand11sectors/S5HLTH INDEX.csv")
s5health
```

```python
s5health['Movement'] = s5health['Close'] - s5health['Open']
s5health
```

```python
vac_sp500health_merge = df_pivot_movement.merge(s5health[['Movement', 'Date']], how="inner", on = "Date")
vac_sp500health_merge
```
