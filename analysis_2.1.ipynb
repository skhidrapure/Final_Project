```python
import pandas as pd
meta = pd.read_csv("datasets/Covid_Companies_Stocks.csv", index_col=False)
meta
```

```python
df = meta[['Ticker', 'Date', 'Close', 'Number of Transactions']]
df.head()
```

```python
df_pivot = df.pivot(index=['Date'], columns = ['Ticker'], values = ['Close','Number of Transactions'])
df_pivot.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')
close = df_pivot.Close
transactions = df_pivot['Number of Transactions']
```

## Overview of closing for vaccination companies

```python
plt.rcParams["figure.figsize"] = (20,6)
close.plot(xlabel = "Date", ylabel = "Closing Price", title = "Closing Price for Vaccination Companies", linewidth = 1.2)
#vac_sp500health_merge.Close.plot()
plt.show()
```

```python
close.reset_index(inplace = True)
close
```

```python
# Loading S5HLTH Index file.

s5health = pd.read_csv("datasets/sp500indexand11sectors/S5HLTH INDEX.csv")
s5health
```

```python
vac_sp500health_merge = close.merge(s5health[['Close','Date']], how="inner", on = "Date")
vac_sp500health_merge
```

```python
fig, ax = plt.subplots()
ax.plot(vac_sp500health_merge.Date, vac_sp500health_merge.iloc[:,1:7], linewidth = 1)
plt.grid(axis = 'x')
plt.xticks([0,360],["2020-03-11","2021-12-30"])

ax2=ax.twinx()
ax2.plot(vac_sp500health_merge.Date, vac_sp500health_merge.Close, color = "black", linestyle = "-.", )
plt.grid()
```

```python
meta_pivot = meta.pivot(index=['Ticker'], columns = ['Date'], values = ['Close'])
meta_pivot.head()
```

```python
meta_pivot_return = meta_pivot.Close[["2020-03-11","2021-12-30"]]
stock_return = (meta_pivot_return["2021-12-30"] - meta_pivot_return["2020-03-11"])/meta_pivot_return["2020-03-11"]*100
meta_pivot_return["Return from 2020-03-11 to 2021-12-30(%)"] = round(stock_return, 2)
```

```python
spxHealth = pd.read_csv("datasets/SPX_Health_Sector_stocks_all.csv")
spxHealth
```

```python
spx_Health_pivot = spxHealth.pivot(index=['Ticker'], columns = ['Date'], values = ['Close'])
spx_Health_pivot
```

```python
spx_health_return = spx_Health_pivot.Close[["2020-03-11","2021-03-11"]]
spx_health_return["Annual Return (%)"] = round((spx_health_return["2021-03-11"] - spx_health_return["2020-03-11"])/spx_health_return["2020-03-11"]*100, 2)
```

```python
spx_health_return = spx_health_return.reset_index()
```
